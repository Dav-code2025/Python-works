```python
#1.The syntax for choosing and loading Penguins dataset
df = sns.load_dataset("penguins")
df
```

```python
#6. Display the number of null values in the dataset and the total count of it
print(df.isna().sum())
print('\nThe total count for the null values is: ', df.isna().sum().sum())
```

```python
#7. Identify the columns having null values and remove them
columns_with_null= df.isna().any()
print(columns_with_null)

# Dropping the with null values
df_cleaned = df.dropna()
df_cleaned
```

```python
# 8.Print the shape of the original and cleaned DataFrames
print('The shape of the original dataframe is ', df.shape)
print('\nThe shape of the cleaned dataframe is ', df_cleaned.shape)
```

```python
#9.Find out the count of male and female penguins
print('The value counts of male and female penguins are;')
df['sex'].value_counts()
```

```python
#10.The syntax to display the count of species 
print('The count values of species are:\n')
print(df['species'].value_counts())
print('\nThe total count of species is:',df['species'].count())
```

```python
#11. Use a countplot graph to display the number of male and female penguins
#Give the plot a title "Count of male and female penguins"
#Give a xlabel and ylabel 
plt.figure(figsize=(8, 6)) 
sns.countplot(data=df_cleaned, x='sex', hue='sex', legend=False)
plt.title('COUNT OF MALE AND FEMALE PINGUINS', fontsize=16)
plt.xlabel('Sex', fontsize=12) 
plt.ylabel('Count', fontsize=12) 
plt.xticks(fontsize=10)  
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#12. For all the penguins display the bill_length_mm and bill_depth_mm using pairplot
sns.pairplot(df_cleaned[['bill_length_mm', 'bill_depth_mm']],diag_kind='hist') 
plt.suptitle("Pair Plot of Bill Dimensions", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
#13.Create a visualization to display the flipper_length_mm and body_mass_g for the penguin dataset.

plt.figure(figsize=(10, 6))
sns.relplot(data=df_cleaned, x='flipper_length_mm', y='body_mass_g', hue='sex', style='sex', markers=['o', 'o'],  
            alpha=0.8)
plt.title('Flipper Length vs. Body Mass by Sex', fontsize=16)
plt.xlabel('Flipper Length (mm)', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
#plt.legend(title='Sex', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#14. Create a histogram with the flipper_length_mm and bins = 10

sns.histplot(data=df_cleaned, x='flipper_length_mm', bins=10)
plt.title('Flipper Length Histogram', fontsize=16)
plt.xlabel('Flipper Length (mm)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
#15. Create a joint point with body_length_mm, body_depth_mm for penguins_cleaned dataset
sns.jointplot(data=df_cleaned, x='bill_length_mm', y='bill_depth_mm',)
plt.show()
```

```python
#16. Use a pairwise plot to display all the numerical values of the dataset penguins_cleaned

sns.pairplot(df_cleaned, diag_kind='hist')
plt.suptitle("Pair Plot of Penguin Data", fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python

#print('The boxplot of body_mass_g of penguins dataset to check outlies')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_cleaned, y='body_mass_g', width=0.5)  
plt.title('Body Mass of Penguins', fontsize=16)
plt.ylabel('Body Mass (g)', fontsize=12)  
plt.yticks(fontsize=10) 

plt.tight_layout()
plt.show()

print('\n>Based on the boxplot above, there are potential outliers in the body_mass_g column of the penguins cleaned dataset.')
```

```python
#print('The boxplot of flipper_length_mm of penguins dataset to check outlies')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_cleaned, y='flipper_length_mm', width=0.5)
plt.title('Flipper Length of Penguins', fontsize=16)
plt.ylabel('Flipper Length (mm)', fontsize=12)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
print('\n>Based on the boxplot above, there are potential outliers in the flipper_length_mm column of the penguins cleaned dataset.')
```

```python
#print('The boxplot of bill_depth_mm of penguins dataset to check outlies')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_cleaned, y='bill_depth_mm', width=0.5)
plt.title('Bill Depth of Penguins', fontsize=16)
plt.ylabel('Bill Depth (mm)', fontsize=12)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
print('\n>Based on the boxplot above, there are potential outliers in the bill_depth_mm column of the penguins cleaned dataset.')
```

```python
#print('The boxplot of bill_length_mm of penguins dataset to check outlies')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_cleaned, y='bill_length_mm', width=0.5)
plt.title('Bill Length of Penguins', fontsize=16)
plt.ylabel('Bill Length (mm)', fontsize=12)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
print('\n>Based on the boxplot above, there are potential outliers in the bill_length_mm column of the penguins cleaned dataset.')
```

```python
#18. Create a violin plot to display the body_mass_g of different species based on sex
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_cleaned, x='species', y='body_mass_g', hue='sex')            
plt.title('Body Mass Distribution by Species and Sex', fontsize=16)  # More descriptive title
plt.xlabel('Species', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Sex', fontsize=10)

plt.tight_layout()
plt.show()
```
